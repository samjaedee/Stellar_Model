# Simple Stellar Model

### Introduction

This project is part of the AST221 course at the University of Toronto. In this project, using our simple density model, we derived simple theoretical equations for density, pressure, and temperature as a function of radius. We also read in the values for density, pressure, and temperature as a function of radius from a full stellar model, in the file "AppendixA2.txt". Finally, we compared those values and examined how well we are able to reproduce the full stellar model with a simple approximation for the density.

### Code

```python
# Import the needed libraries and packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import astropy.constants as c
import astropy.units as u
```

```python
# Read in the data from AppendixA2 and separate all the columns into different arrays

norm_mass, norm_radius, temperature, density, pressure = np.loadtxt('AppendixA2.txt', skiprows = 26, usecols = (0, 1, 2, 3, 4), unpack = True)
```

```python
# Let's look at the first five values of the normalized mass and the normalized radius

print("The first five values of the normalized mass are: " + str(norm_mass[:5]))
print("The first five values of the normalized radius are: " + str(norm_radius[:5]))
```

```python
# It is important to note that from the dataset, mass is in solar masses, radius is in solar radius, temperature is in Kelvin,
# density is in units of g/cm³, and pressure is in units of dyn/cm²

# Set up some constants, like gravitational constant G, the solar mass M_sun, the solar radius R_sun, mass of hydrogen atom M_H,
# Boltzmann constant k, and mean particle mass of the sun mu = 0.61

G = c.G
M_sun = c.M_sun
R_sun = c.R_sun
M_H = c.m_p
k = c.k_B
mu = 0.61

# In this cell, convert the temperature, density, and pressure to SI units using this conversion
# 1 dyne = 1 g cm/s²
# 1 dyne/cm² = 1 N/m² = 0.1 Pa
# 1 g/cm³ = 1000 kg/m³

temperature = temperature * u.Kelvin
density = density * 1000 * (u.kilogram / u.meter ** 3)
pressure = pressure * 0.1 * u.Pa

# Also define a parameter, that is, the central density using this formula

c_density = (3 * M_sun) / (np.pi * R_sun ** 3)
```

The equation to find the density for all radius is given by $\rho(r) = \rho_c (1 - r/R)$ where $\rho_c$ is the central density and $r/R$ is the normalized radius. Since our radius array is normalized with $x = r/R$, we can rewrite this as $\rho(x) = \rho_c (1 - x)$. Here, create an array for $\rho(r)$.

```python
r_density = c_density * (1 - norm_radius)

# Create a plot of the density against the normalized radius

fig = plt.figure()

plt.plot(norm_radius, r_density, color = 'black')

plt.xlabel('Normalized radius ($r/R$)')
plt.ylabel('Density ($kg/m^3$)')
plt.title('Plot of Density vs Normalized Radius')

plt.show()
```

```python
# Then, using the known equations, make an array for pressure and an array for temperature as a function of the normalized radius

r_pressure = (np.pi * G * (r_density ** 2) * (R_sun ** 2)) * ((5/36) + ((-2/3) * norm_radius ** 2) + ((7/9) * norm_radius ** 3) - ((1/4) * norm_radius ** 4))
r_temperature = ((r_pressure * mu * M_H) / (r_density * k)).decompose()

# To check, print the first five values of r_pressure and r_temperature

print('First five values of pressure: ')
print(r_pressure.value[0:5])
print('First five values of temperature: ')
print(r_temperature.value[0:5])
```

Next, plot and compare the data of density, pressure, and temperature between the observational data and the theoretical calculations.

```python
# Plot and compare the density

fig = plt.figure()

plt.plot(norm_radius, r_density, color = 'black', label = 'Simple Model') 
plt.plot(norm_radius, density, color = 'red', label = "Actual Observation")

plt.xlabel('Normalized radius (r/R)')
plt.ylabel('Density (kg/m$^3$)')
plt.title('Plot of Density vs Normalized Radius from Simple Model and Actual Observation')
plt.xscale('log')
plt.yscale('log')

plt.legend()
plt.savefig('Density_Plot.png')
```

```python
# Plot and compare the pressure

fig = plt.figure()

plt.plot(norm_radius, r_pressure, color = 'black', label = 'Simple Model') 
plt.plot(norm_radius, pressure, color = 'red', label = "Actual Observation")

plt.xlabel('Normalized radius (r/R)')
plt.ylabel('Pressure (Pa)')
plt.title('Plot of Pressure vs Normalized Radius from Simple Model and Actual Observation')
plt.xscale('log')
plt.yscale('log')

plt.legend()
plt.savefig('Pressure_Plot.png')
```

```python
# Plot and compare the temperature

fig = plt.figure()

plt.plot(norm_radius, r_temperature, color = 'black', label = 'Simple Model') 
plt.plot(norm_radius, temperature, color = 'red', label = "Actual Observation")

plt.xlabel('Normalized radius (r/R)')
plt.ylabel('Temperature (Pa)')
plt.title('Plot of Temperature vs Normalized Radius from Simple Model and Actual Observation')
plt.xscale('log')
plt.yscale('log')

plt.legend()
plt.savefig('Temperature_Plot.png')
```

### Conclusion

From the three graphs that we see, our simple stellar model is generally within an order of the actual observation. It means that we are successful in approximating the full stellar model with our simple stellar density model. This is remarkable considering our simple assumption of the central density that does not include energy, opacity, etc. Again, the actual observation is well approximated by the simple stellar model.
